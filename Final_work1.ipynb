```python
import pandas as pd
```

```python
data1 =  pd.read_csv('/content/clean_sales_fabric_96_sku_265_warehouse_1.csv')
data2 =  pd.read_csv('/content/clean_sales_fabric_42_sku_13653_warehouse_-1.csv')
```

```python
# Агрегация данных по дате
data1_aggregated = data1.groupby('date').sum().reset_index()
data2_aggregated = data2.groupby('date').sum().reset_index()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(pd.to_datetime(data1_aggregated['date']), data1_aggregated['count'], label='Data 1')
plt.plot(pd.to_datetime(data2_aggregated['date']), data2_aggregated['count'], label='Data 2')
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Прогнозирование для Data 1
model1 = ARIMA(data1_aggregated['count'], order=(5,1,0))
model1_fit = model1.fit()
forecast1 = model1_fit.forecast(steps=40)

# Прогнозирование для Data 2
model2 = ARIMA(data2_aggregated['count'], order=(5,1,0))
model2_fit = model2.fit()
forecast2 = model2_fit.forecast(steps=30)

print("Forecast for Data 1:", forecast1)
print("Forecast for Data 2:", forecast2)
```
